```python
import torch
from PIL import Image
from torchvision import transforms

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/Users/jashtandel/SEM6/UAS/COMP_AGRI/yolov5/runs/train/exp/weights/best.pt')
model.conf = 0.25  # confidence threshold (0-1)
model.iou = 0.10
# Load image
image_path = '/Users/jashtandel/SEM6/Train_data/images/train/7xm.xyz889804.jpg'
img = Image.open(image_path)

# Define transforms
transform = transforms.Compose([
    transforms.Resize((640, 640)),
    transforms.ToTensor(),
])

# Apply transforms to image
img_tensor = transform(img).unsqueeze(0)  # Add batch dimension

# Perform inference with thresholding value
results = model(img_tensor)  # Adjust the threshold and IoU value as needed

# Print results
results.print()
results.xyxy[0]
results.pandas().xyxy[0]
```

```python
import numpy as np

re = results.pandas().xyxy
total = np.shape(re)[0]    # total number of images
print('total images =', total)

unf = np.array([])     # number of unfrozen droplets 
sl = np.array([])     # slices / #images 

c = 1
while c <= total:
   
   for i in re:
           
       n = np.shape(i)[0]   # number of detected objects in each slice/image 
       
       if n > 0:
           
           unf = np.append(unf,n)
           sl = np.append(sl,c)
           
       c += 1
                                      
print(unf) 
print (sl)  
```
